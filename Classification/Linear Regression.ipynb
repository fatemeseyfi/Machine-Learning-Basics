```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("student-mat.csv", sep=";")
df
```

```python
df.columns
```

```python
df = df[['G1', 'G2', 'G3', 'studytime','failures', 'absences']]
df
```

```python
#       *** split features and lables ***

predict = 'G3'
# features of data:
X = np.array(df.drop(columns=predict, axis = 1))
# labels of data:
y = np.array(df[predict])
```

```python
# test set and training set:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
# accuracy:
acc = linear_model.score(X_test, y_test)
acc
```

```python
# coefficents and intercept:  y = mx + b
print("co: ", linear_model.coef_)  # m
print("intercept: ", linear_model.intercept_) # b
```
