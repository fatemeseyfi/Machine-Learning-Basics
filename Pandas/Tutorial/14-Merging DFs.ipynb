```python
import numpy as np
import pandas as pd
```

```python
data1 = {
  'id': [1,2,3,4,5],
  'city':['Tripoli', 'London', 'Rome', 'Cairo', 'Sydney'],
  'rank': ['1st', '2nd', '3rd', '4th', '5th'],
  'population': ['1m','2m','3m','4m','5m']
}
df1 = pd.DataFrame(data1)
```

```python
data2 = {
  'id': [6,7,8,9],
  'city':['Lisbob', 'Tokyo', 'Mecca', 'Berlin'],
  'rank': ['19th', '22nd', '33rd', '40th'],
  'population': ['1m','9m','0.8m','7m']
}
df2 = pd.DataFrame(data2)
df2
```

```python
data3 = {
  'id': [1,4,8,9],
  'city':['Milano', 'Bejing', 'Addis Ababa', 'Belfast'],
  'rank': ['7th','6th','21st','11th']
}
df3 = pd.DataFrame(data3)
df3
```

```python
# merge df2 and df3
pd.merge(df2, df3, on='id')
```

```python
# merge df1 and df2 --> by default how='inner'
pd.merge(df1, df2, on='id')
```

```python
# merge df2 and df3
pd.merge(df2, df3, on='id', how='left')
```

```python
# merge df2 and df3
pd.merge(df2, df3, on='id', how='right')
```

```python
# merge df2 and df3 everything
pd.merge(df2, df3, on='id', how='outer', indicator=True)
```

```python
pd.merge(df1, df2, on=['id','city'], how='outer', indicator=True)
```

```python
pd.merge(df1, df2, on=['id','city'], how='inner')
```

```python
# change siffixes
pd.merge(df2, df3,on='id', how='left', suffixes=['_1', '_2'])
```
