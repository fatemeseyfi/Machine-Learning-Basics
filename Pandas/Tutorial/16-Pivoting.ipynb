```python
import pandas as pd
```

```python
pp = {'city' : ['Tripoli','Sydney','Tripoli','Rome','Rome', 'Tripoli', 'Rome', 'Sydney', 'Sydney'],
'rank': ['1st','2nd','1st','2nd','1st', '2nd', '1st', '2nd','1st'],
'score1' : [44,48,39,41,38,44,34,54,61],
'score2': [67,63,55,70,64,77,45,66,72],
}

df = pd.DataFrame(pp, columns=['city', 'rank',  'score1', 'score2'])
df
```

```python
# create pivot table of group means, by city and rank
pd.pivot_table(df, index=['city', 'rank'], aggfunc='mean')
```

```python
# create pivot table of group means, by city and rank
pd.pivot_table(df, index=['city', 'rank'], aggfunc='count')
```
