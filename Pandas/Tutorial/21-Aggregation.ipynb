```python
import numpy as np
import pandas as pd
```

```python
pp = {'city' : ['Tripoli','Sydney','Tripoli','Rome','Rome', 'Tripoli', 'Rome', 'Sydney', 'Sydney'],
'rank': ['1st','2nd','1st','2nd','1st', '2nd', '1st', '2nd','1st'],
'score1' : [44,48,44,41,38,44,34,54,61],
'score2': [67,63,67,70,64,77,45,66,72],
}

df = pd.DataFrame(pp, columns=['city', 'rank',  'score1', 'score2'])
df
```

```python
city_rank = df.groupby(['city', 'rank'])
```

```python
# apply mean function on score 1 and 2 in specific group
city_rank.agg('mean')
```

```python
# apply max, min and mean function on score 1 and 2 in specific group
city_rank.agg(['max', 'min', 'mean'])
```

```python
# a func that finds the mean
def find_mean(v):
  return np.mean(v)
```

```python
# find maen of score1 and score2 with defined func
city_rank.agg(find_mean)
```

```python
# find min, max and maen of score1 and score2 with defined func
city_rank.agg(['min', 'max', find_mean])
```
