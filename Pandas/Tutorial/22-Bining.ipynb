```python
import pandas as pd
```

```python
pp = {'city' : ['Tripoli','Sydney','Tripoli','Rome','Rome', 'Tripoli', 'Rome', 'Sydney', 'Sydney'],
'rank': ['1st','2nd','1st','2nd','1st', '2nd', '1st', '2nd','1st'],
'score1' : [44,48,44,41,38,44,34,54,61],
'score2': [50,63,67,70,64,75,45,66,72],
}

df = pd.DataFrame(pp, columns=['city', 'rank',  'score1', 'score2'])
df
```

```python
# define bins as 0-25, 25-50, 50-75, 75-100
bins = [0,25,50,75,100]
```

```python
# name for the four groups
group_names = ['low', 'average', 'good', 'brilliant']
```

```python
df['grade'] = pd.cut(df['score2'], bins, labels=group_names)
df
```
