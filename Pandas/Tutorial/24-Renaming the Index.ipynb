```python
import pandas as pd
```

```python
pp = {'city' : ['Tripoli','Sydney','Tripoli','Rome','Rome', 'Tripoli', 'Rome', 'Sydney', 'Sydney'],
'rank': ['1st','2nd','1st','2nd','1st', '2nd', '1st', '2nd','1st'],
'score1' : [44,48,44,41,38,44,34,54,61],
'score2': [67,63,67,70,64,77,45,66,72],
}

df = pd.DataFrame(pp,
                    index= pd.Index(['A', 'B','C','D', 'E', 'F', 'G','H','I'], name='letters'),
                    columns=pd.Index(['city', 'rank',  'score1', 'score2'], name='attributes'))
df
```

```python
# rename an index
df.rename(index={'A':'a'})
```

```python
# rename a column
df.rename(columns={'rank':'Rank'})
```

```python
# rename cols:
df.columns=['City', 'Rank', 'Score1', 'Score2']
df
```

```python
df.columns = [x.upper() for x in df.columns]
df
```

```python
df.columns = df.columns.map(lambda x: x.upper())
df
```

```python
df.columns = df.columns.map(lambda x: x.lower())
df
```

```python
df.rename(index={'D': 'dd', 'G':'g'}, columns={'rank':'level'}, inplace=True)
df
```
