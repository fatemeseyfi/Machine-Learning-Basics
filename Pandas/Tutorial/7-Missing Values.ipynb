```python
import pandas as pd
import numpy as np
```

```python

pp = {'first_name' : ['Josef','Mike','Luke','Noureddin','Ali'],
'last_name': ['Baker','Brooks','Brown','Sadawi','Shah'],
'score' : [322,233,401,422,223],
'year': [2009,2013,2014,2015,2016],

}
df = pd.DataFrame(pp, index = ['Birminghom', 'London', 'New York','Tripoli','Islamabad'])
df
```

```python
flt1 = (df['score'] > 240 )
flt2 = (df['year'] > 240 )
df.loc[flt1, 'score'] = np.NaN
df.loc[flt2, 'year'] = np.nan
df
```

```python
# how = any -> remove the row if any of its columns are null
# how = all -> remove the row if all of its columns are null
df.dropna(how='all') # -> will drop none of rows
```

```python
df.dropna(how='any') # -> will drop all of rows
```

```python
# by default axis=0
# axis = 0 --> of rows
# axis = 1 --> of cols
df.dropna(axis = 1, how='any')
```

```python
df.dropna(axis=1, how='all')
```

```python
# which are null
df.isnull()
```

```python
# fill null with other character
df.fillna('_')
```
