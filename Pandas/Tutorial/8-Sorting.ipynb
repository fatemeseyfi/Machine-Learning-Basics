```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(20).reshape(5,4), index=['r1', 'r2', 'r3', 'r4','r5'])
df.columns = ['c1', 'c2', 'c3', 'c4']
df
```

```python
# sort by values of one col
df.sort_values(by='c2')
```

```python
# sort by values of 2 cols
df.sort_values(by=['c1', 'c2'])
```

```python
pp = {'first_name' : ['Josef','Mike','Luke','Noureddin','Ali'],
'last_name': ['Baker','Brooks','Brown','Sadawi','Shah'],
'score' : [322,233,422,422,223],
'year': [2009,2013,2014,2015,2016],

}
df = pd.DataFrame(pp, index = ['Birminghom', 'London', 'New York','Tripoli','Islamabad'])
df
```

```python
df.sort_values(by='score', ascending=True)
```

```python
df.sort_values(by='score', ascending=False)
```

```python
df.sort_values(by=['score', 'first_name'], ascending=[False, True])
```

```python
df.sort_index(ascending=False)
```
