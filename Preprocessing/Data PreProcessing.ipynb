Value Imputation

```python
from sklearn.impute import SimpleImputer
```

```python
# number of null values in 'age' column before imputation
df['age'].isnull().sum()
```

```python
# SimpleImputer replace the null values in 'age' col with average value of non null values
imp = SimpleImputer(strategy='mean')
df['age'] = imp.fit_transform(df[['age']])
# number of null values in 'age' column after imputation
df['age'].isnull().sum()
```

```python
df.isnull().any()
```

```python
# columns that contain null value
flt = df.isnull().any()
df.columns[flt]
```

```python
for col in df.columns[df.isnull().any()]:
  print(df[col][df[col].isnull()])
```

```python
# select proper strategy for each column that contains null value
def get_parameters(df):
  parameters={}
  for col in df.columns[df.isnull().any()]:

    if df[col].dtype in ['float64', 'int64', 'int32']:
      strategy='mean'
    else:
      strategy='most_frequent'

    missing_values = df[col][df[col].isnull()].values[0]
    parameters[col] = {'missing_values':missing_values, 'strategy':strategy}
  return parameters
```

```python
parameters = get_parameters(df)
parameters
```

```python
for col, param in parameters.items():
  print(col)
```

```python
import numpy as np
```

```python
def imputer(df, parameters):
  for col, param in parameters.items():
    stg = param['strategy']
    imp = SimpleImputer(strategy=stg)
    df[col] = imp.fit_transform(df[[col]]).ravel()
```

```python
imputer(df, parameters)
```
